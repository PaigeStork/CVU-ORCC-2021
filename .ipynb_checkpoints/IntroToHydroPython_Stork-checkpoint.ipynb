```python
import numpy
import matplotlib.pyplot as plt
import pandas as pd
from nwis import Nwis
```

```python
site_id = '13185000' #USGS HUC code for Boise River @ Twin Springs
my_start = '2015-10-01' #start date for data analysis
my_end = '2020-09-30' #end date for data analysis
```

```python
#Get data from NWIS
nwis_data = Nwis()
dataset = nwis_data.get_data(site=site_id, start_date=my_start, end_date=my_end, data_type='dv')
```

```python
dataset
```

```python
dataset['00060']
```

```python
Q = dataset['00060'].values
dates = dataset['datetime'].values
Q.shape
```

```python
plt.plot(dates,Q)
```

```python
plt.figure(figsize=(14,10))
plt.rcParams.update({'font.size':18})
plt.plot(dates,Q)
plt.grid()
plt.title('Boise River at Twin Springs ID')
plt.xlabel('Date')
plt.ylabel('Discharge [ft^3/s]')
plt.show()
plt.savefig('Hydrograph_BRTS.png',dpi=300)
```
