```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt

#ncfile = '../downloads/c1184286e03e4f4ba38d65bd5754beb2/c1184286e03e4f4ba38d65bd5754beb2/data/contents/WRF-NAAR-1km-WY2017/WRF-NARR-1km-WY2017.nc'

ncfile = '~/downloads/c1184286e03e4f4ba38d65bd5754beb2/c1184286e03e4f4ba38d65bd5754beb2/data/contents/WRF-NARR-1km-WY2017/WRF-NARR-1km-WY2017.nc'
ncfile
```

```python
ds = xr.open_dataset(ncfile)
ds
```

```python
da_seasonal_prcp = ds['DPRCP'].groupby('XTIME.month').sum(dim='XTIME') #groups the data, then sums it
da_seasonal_prcp
```

```python
da_seasonal_prcp.plot(x='west_east', y='south_north', col='month', col_wrap=3, cmap='cividis_r') #facet grid/plot; wrap every third column
```

```python
da_seasonal_anom = da_seasonal_prcp - ds['DPRCP'].sum(dim='XTIME') / 12
da_seasonal_anom
```

```python
da_seasonal_anom.plot(x='west_east', y='south_north', col='month', col_wrap=3, cmap='Spectral')
```
